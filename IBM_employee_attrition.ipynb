```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
import seaborn as sns
df1 = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
df=df1.copy()
data = df.drop(["EmployeeCount", "StandardHours",'Over18'], axis=1)
#data.head()
data['Attrition'] =data['Attrition'].astype('category').cat.codes #to directly convert categorical data to numerical data
data['BusinessTravel'] =data['BusinessTravel'].astype('category').cat.codes   #yes
data['Department'] =data['Department'].astype('category').cat.codes  #yes
data['Gender'] =data['Gender'].astype('category').cat.codes         #yes
data['JobRole'] =data['JobRole'].astype('category').cat.codes       #yes
data['MaritalStatus'] =data['MaritalStatus'].astype('category').cat.codes   #yes
data['OverTime'] =data['OverTime'].astype('category').cat.codes         #yes , very much significant
data['EducationField'] =data['EducationField'].astype('category').cat.codes    #yes

data1=data.drop(['Attrition','BusinessTravel','Department','Gender','JobRole','MaritalStatus','OverTime','EducationField'],axis=1)#removing categorical data for making correlation heatmap
plt.figure(figsize=(10,10))
cor = data1.corr()
ax = sns.heatmap(cor, vmin=-1, vmax=1, center=0,cmap='YlGnBu',square=True)
ax.set_xticklabels(ax.get_xticklabels(),rotation=45,horizontalalignment='right')
plt.show()
```
